```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./smartphones.csv')
```

```python
null_cols=[]
df.columns
for i in df.columns:
    if any(df[i].isna()):
        null_cols.append(i)
```

```python
null_cols
```

```python
df[null_cols].head()
```

```python
df_missing=df[null_cols]
```

```python
df_missing.describe().style.background_gradient(cmap='Oranges')
```

```python
mean_rating=df['rating'].mean()
df['rating'].fillna(mean_rating,inplace=True)
```

```python
df['rating'].isna().sum()
```

```python
df['fast_charging'].unique()
```

```python
df['fast_charging'].describe()
```

```python
charging_values=[20,25,40,55,65]
random_values=np.random.choice(charging_values,df['fast_charging'].isna().sum(),replace=True)
random_values
```

```python
df.loc[df['fast_charging'].isna(),'fast_charging']=random_values
```

```python
df['fast_charging'].isna().sum()
```

```python
df['extended_upto'].fillna(df['extended_upto'].mean(),inplace=True)
df['battery_capacity'].fillna(df['battery_capacity'].mean(),inplace=True)
df['processor_speed'].fillna(df['processor_speed'].mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.describe().T.style.background_gradient(cmap='plasma')
```

```python
num_cols=df.select_dtypes(include=['int64','float64']).columns
```

```python
num_cols
```

```python
fig,axis=plt.subplots(7,3,figsize=(20,35))
axis=axis.flatten()
for i,col in enumerate(num_cols):
    sns.histplot(data=df,x=col,kde=True,color='blue',ax=axis[i])
    title = col.replace('_', ' ').title()  
    axis[i].set_title(title)


plt.tight_layout()
plt.show()    
```

```python
os_counts=df['os'].value_counts()
subplot_color = 'white'
fast_charging_counts = df['fast_charging_available'].value_counts()
fig, axes = plt.subplots(2, 2, figsize=(10, 8), facecolor=subplot_color)
sns.countplot(data=df, x='os', ax=axes[0, 0])
axes[0, 0].set_xlabel('OS')
axes[0, 0].set_ylabel('Count')
axes[0, 0].set_facecolor(subplot_color)


axes[0, 1].pie(os_counts.values, labels=os_counts.index, autopct='%1.2f%%')
axes[0, 1].set_facecolor(subplot_color)

axes[0, 0].set_title('Distribution of Operating Systems')
axes[0, 1].set_title('Distribution of Operating Systems')

sns.countplot(data=df, x='fast_charging_available', ax=axes[1, 0])
axes[1, 0].set_xlabel('Fast Charging Available')
axes[1, 0].set_ylabel('Count')
axes[1, 0].set_facecolor(subplot_color)


axes[1, 1].pie(fast_charging_counts.values, labels=fast_charging_counts.index, autopct='%1.2f%%')
axes[1, 1].set_facecolor(subplot_color)

axes[1, 0].set_title('Distribution of Fast Charging Availability')
axes[1, 1].set_title('Distribution of Fast Charging Availability')

plt.tight_layout()

plt.show()
```

```python
brands=df['brand_name'].value_counts()
brands_index=brands.index
plt.figure(figsize=(14,6))
sns.countplot(data=df,x='brand_name',order=brands_index)
plt.xticks(rotation=90)
plt.show()
```

```python
df.info()
```

```python
bool_cols=['has_5g','has_nfc','has_ir_blaster']
plt.figure(figsize=(16,10))

for i,feature in enumerate(bool_cols,1):
    counts=df[feature].value_counts()
    plt.subplot(2, 3, i)
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab10.colors)
    plt.title(f'{feature.capitalize()} Distribution')
    plt.axis('equal')
    
plt.tight_layout()
plt.show()
    
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64'])
corr_matrix = numeric_cols.corr()
plt.figure(figsize=(20, 12))
sns.heatmap(corr_matrix, annot=True, cmap='tab20c', fmt=".2f", vmin = -1, vmax = 1)
plt.title('CORRELATION HEAT MAP OF MOBILE FEATURES')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x=df['num_front_cameras'],y=df['num_rear_cameras'])
plt.title('Font cameras vs Back cameras')
plt.xlabel('Font cameras')
plt.ylabel('Back cameras')
plt.show()
```

```python
sorted_df=df.sort_values(by='brand_name',ascending=True)
plt.figure(figsize=(6,4))
sns.barplot(data=sorted_df,x='brand_name',y='rating')
plt.title('BRAND VS RATING')
plt.xlabel('Brand')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.barplot(data=sorted_df,x='brand_name',y='price')
plt.xticks(rotation=90)
plt.show()
```

```python

plt.figure(figsize=(6,4))
sns.barplot(data=sorted_df,x='processor_brand',y='price')
plt.title("Processor brand vs Price")
plt.xticks(rotation=90)
plt.show()
```
